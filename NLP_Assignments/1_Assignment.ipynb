# Traditional way to findout special character are present into data.
-  so we will use regex to clean the data from special character.

```python
# Create an empty set to store unique special characters
special_characters = set()

# Iterate through each character in the text_data
for char in text_data:
    # Check if the character is not alphanumeric
    if not char.isalnum():
        # Add it to the set of special characters
        special_characters.add(char)

# Convert the set of special characters to a sorted list
special_characters_list = sorted(list(special_characters))

# Print the list of special characters
print("Special characters in the text_data:")
for char in special_characters_list:
    print(char)
```

# regex

```python
# Define a regex pattern to match the specified special characters
special_characters_pattern = r'[!"\-,./:‚ù§Ô∏èüíÄüòÇüòäüò≠ü•∞]'

# Remove the specified special characters from the text_data
cleaned_text_data = re.sub(special_characters_pattern, '', text_data)

# Print the cleaned text data
cleaned_text_data
```
